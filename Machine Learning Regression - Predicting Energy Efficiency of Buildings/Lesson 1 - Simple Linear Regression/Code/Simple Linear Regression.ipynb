```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('dataset/energy_efficiency.xlsx')
df
```

```python
# renaming columns
column_names = {
    'X1':'Relative_Compactness','X2':'Surface_Area',
    'X3': 'Wall_Area', 'X4': 'Roof_Area',
    'X5': 'Overall_Height', 'X6': 'Orientation',
    'X7': 'Glazing_Area', 'X8': 'Glazing_Area_Distribution',
    'Y1': 'Heating_Load', 'Y2': 'Cooling_Load'
}
df = df.rename(columns = column_names)
df
```

```python
# sampling our dataset
simple_linear_regression_df = df[['Relative_Compactness','Cooling_Load']].sample(10, random_state=2)
simple_linear_regression_df
```

```python
# regression plot
sns.regplot(x='Relative_Compactness', y='Cooling_Load', data=simple_linear_regression_df)
```
