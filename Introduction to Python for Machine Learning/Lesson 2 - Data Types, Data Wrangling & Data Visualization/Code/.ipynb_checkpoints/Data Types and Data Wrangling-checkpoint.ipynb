```python
# reading urls
url = "https://github.com/WalePhenomenon/climate_change/blob/master/fuel_ferc1.csv?raw=true"
fuel_data = pd.read_csv(url, error_bad_lines=False)
fuel_data.describe(include='all')
```

```python
# checking for the sum of missing values
fuel_data.isnull().sum()
```

```python
# using groupby to count the sum of each unique value in the fuel unit column OR we are grouping our dataframe by the fuel_unit column
fuel_data.groupby('fuel_unit')['fuel_unit'].count()
# filling null/missing values with mcf
fuel_data['fuel_unit'] = fuel_data['fuel_unit'].fillna(value='mcf')
# check if missing values have been filled
fuel_data.isnull().sum()
# groupby the report year and count each value for each year
fuel_data.groupby('report_year')['report_year'].count()
# groupby the fuel_type_code_pudl and print the first value
fuel_data.groupby('fuel_type_code_pudl').first()
```

```python
fuel_df1 = fuel_data.iloc[0:19001]
fuel_df2 = fuel_data.iloc[19001:]
# Checking the length of both df are equal to the expected length
assert len(fuel_data) == (len(fuel_df1) + len(fuel_df2))
```

```python
# Inner merge
fuel_df1_df2_inner = pd.merge(fuel_df1, fuel_df2, how='inner')
fuel_df1_df2_inner
```

```python
# Outer merge
fuel_df1_df2_outer = pd.merge(fuel_df1, fuel_df2, how='outer')
print("Outer merge")
fuel_df1_df2_outer
```

```python
# left merge
fuel_df1_df2_left = pd.merge(fuel_df1, fuel_df2, how='left')
print("Left merge")
fuel_df1_df2_left
```

```python
# right merge
fuel_df1_df2_right = pd.merge(fuel_df1, fuel_df2, how='right')
print('Right Merge')
fuel_df1_df2_right
```
